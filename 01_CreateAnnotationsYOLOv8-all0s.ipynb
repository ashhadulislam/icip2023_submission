### Find all classes

```python
all_xml_files=os.listdir("AllAnnotations/")
label_set=set()
for f in all_xml_files:
    if ".DS" in f:
        continue
    print(f)
    content = []
    # Read the XML file
    with open("AllAnnotations/"+f, "r") as file:
        # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = bs(content, "lxml")
    labels_list=bs_content.find_all("labels")
    for labels in labels_list:
        ls=labels.find_all("label")
        for l in ls:
            label_name=l.find("name").text      
            label_set.add(label_name)


            
```

```python
label_set_list=list(label_set)
label_set_list.sort()
print(label_set_list)
```

```python
dic_labels={}
for i in range(len(label_set_list)):
    dic_labels[label_set_list[i]]=i
    
print(dic_labels)    
```

### Find image sizes

```python
all_xml_files=os.listdir("AllAnnotations/")
label_set=set()
for f in all_xml_files:
    if ".DS" in f:
        continue
    content = []
    # Read the XML file
    with open("AllAnnotations/"+f, "r") as file:
        # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = bs(content, "lxml")
    width=int(bs_content.find("original_size").find("width").text)
    height=int(bs_content.find("original_size").find("height").text)
    print(f,width,height)
```

```python
annotation_dic={}

'''
This is what it should look like
annotation_dic"{
    file_name:{
        frame_num":
            [
                [class,x,y,h,w],
                [class,x,y,h,w],                
            ]
        
    }
}


'''
```

```python
annotation_dic={}
for f in all_xml_files:
    if ".DS" in f:
        continue
    print(f)
    annotation_dic[f]={}
    
    
    content = []
    # Read the XML file
    with open("AllAnnotations/"+f, "r") as file:
        # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = bs(content, "lxml")
    width=int(bs_content.find("original_size").find("width").text)
    height=int(bs_content.find("original_size").find("height").text)
    print(f,width,height)    
    all_tracks = bs_content.find_all("track")    
    
    for track in all_tracks:   
        label=track.get("label")
        cls=dic_labels[label]
        tracking_id=track.get("id")
        all_boxes=track.find_all("box")
        for box in all_boxes:
            xtl=float(box.get("xtl"))
            xbr=float(box.get("xbr") )    
            ytl=float(box.get("ytl"))
            ybr=float(box.get("ybr"))
            frame_num=int(box.get("frame"))
            if frame_num not in annotation_dic[f]:
                annotation_dic[f][frame_num]=[]
            
#             print(tracking_id,"class",cls,label,"present in frame",frame_num,"at top left (x,y):",xtl,ytl,"bottom right (x,y)",xbr,ybr)
            center_x=(xbr+xtl)/2/width
            center_y=(ybr+ytl)/2/height
            width_prop=(xbr-xtl)/width
            height_prop=(ybr-ytl)/height
            str_line=str(cls)+" "+str(center_x)+" "+str(center_y)+" "+str(width_prop)+" "+str(height_prop)
            # below line for the special case whr each class is 0
#             str_line=str(0)+" "+str(center_x)+" "+str(center_y)+" "+str(width_prop)+" "+str(height_prop)
            annotation_dic[f][frame_num].append(str_line)

    
    
#     break
```

### Separate folders for each class

```python
'''
As we need separate models for each class, we need to create separate folders for each

Folder structure can be as follows

yolov8_annotations/
    bottle/
        video_1_frame_x.txt
        video_1_frame_y.txt        
        ...
        video_1_frame_n.txt        
        video2_frame_k.txt
        ...
        videon_frame_l.txt
    bucket/
        video_1_frame_x.txt
        video_1_frame_y.txt        
        ...
        video_1_frame_n.txt        
        video2_frame_k.txt
        ...
        videon_frame_l.txt

        
    

'''
```

```python
# lets create the folders
src="yolov8_annotations_every_class_is_0"
if not os.path.isdir(src):
    os.mkdir(src)
    
for k in list(dic_labels.keys()):   
    if not os.path.isdir(src+"/"+k):
        os.mkdir(src+"/"+k)    
    
```

```python
# time to store the dictionary as annotations
```

```python
for class_labell in os.listdir(src):
    if ".DS_" in class_labell:
        continue
    class_value=dic_labels[class_labell]
    print("For class",class_labell,"having value ",class_value)
    for vid_file in annotation_dic.keys():
        only_vid_name=vid_file.split(".")[0]
        print(vid_file,only_vid_name)
        this_vid_details=annotation_dic[vid_file]
        for frame_number in this_vid_details.keys():
            
            this_frame_details=this_vid_details[frame_number]
#             print("Frame number ",frame_number,"has ",len(this_frame_details),"lines")
            # read all the lines and take it if it matches with the class under scrutiny
            for this_line in this_frame_details:
                components=this_line.split(" ")

                if str(class_value)==str(components[0]):
#                     print("matched with",components[0])
                    # now add it to the file
                    # lets see what the file name will be
                    target_file_name=src+"/"+class_labell+"/"+str(only_vid_name)+"_"+str(frame_number)+".txt"
#                     print("Let us write to ",target_file_name)
                    with open(target_file_name, 'a+') as f:
                        new_this_line="0"+this_line[1:]
                        
#                         print("writing",this_line)
                        f.write(new_this_line+"\n")
    
```
